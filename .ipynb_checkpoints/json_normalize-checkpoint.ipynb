```python
import pandas as pd
```

```python
data = pd.read_json('client_data.json')
```

```python
dictionary = {}
for i in data['_source'][0].keys():
    if i not in dictionary:
        dictionary[i] = []
for i in range(len(data)):
    key = data['_source'][i].keys()
    for j in key:
        dictionary[j].append(data['_source'][i][j])
```

```python
df2 = pd.DataFrame(dictionary)
```

```python
df2
```

```python
df = pd.concat([data,df2],axis=1)
```
